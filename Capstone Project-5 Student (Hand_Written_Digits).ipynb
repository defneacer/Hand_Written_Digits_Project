# Support Vector Machine (SVM) Classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

```python
X = df.drop('class', axis=1)
y = df['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, stratify=y, random_state=61)
```

```python
model = SVC(random_state=61)
model.fit(X_train, y_train) 
```

```python
eval_metric(model, X_train, y_train, X_test, y_test)
```

```python
from sklearn.model_selection import cross_validate
```

```python
model = SVC(random_state=61)

scores = cross_validate(model, X_train, y_train, scoring=["accuracy", "recall_macro", "precision_macro", "f1_macro"], cv=10, return_train_score=True)
df_scores = pd.DataFrame(scores, index=range(1, 11))

df_scores.mean()[2:]
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.pipeline import Pipeline
```

```python
steps = [('scaler', MinMaxScaler()), ('SVC', SVC(random_state=61))]
pipe_model = Pipeline(steps)
```

```python
pipe_model.fit(X_train, y_train)
```

```python
scores = cross_validate(pipe_model, X_train, y_train, scoring=["accuracy", "recall_macro", "precision_macro", "f1_macro"], cv=10, return_train_score=True)
df_scores = pd.DataFrame(scores, index=range(1, 11))

df_scores.mean()[2:]
```

## Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
svm_grid_model.best_params_
```

# Decision Tree (DT) Classifier

```python
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor


operations = [("scaler", StandardScaler()), ("DT_model", DecisionTreeClassifier(random_state=61))]

pipe_model = Pipeline(steps=operations)

pipe_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
ConfusionMatrixDisplay.from_estimator(pipe_model, X_test, y_test)
```

```python
eval_metric(pipe_model, X_train, y_train, X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score
from sklearn.metrics import make_scorer
```

```python
scores = cross_validate(pipe_model, X_train, y_train, scoring = ["accuracy", "precision_micro", "recall_micro", "f1_micro"], 
                        cv = 10, return_train_score=True)
df_scores = pd.DataFrame(scores, index = range(1, 11))
df_scores
```

```python
df_scores.mean()[2:] #overfitting var!!
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from itertools import cycle

pipe_model.fit(X_train, y_train)

# Hedef değişkeni ikili formata dönüştürüyoruz
y_test_binarized = label_binarize(y_test, classes=np.unique(y_test))
n_classes = y_test_binarized.shape[1]

# Tahmin olasılıklarını alıyoruz
y_pred_proba = pipe_model.predict_proba(X_test)

# Her sınıf için ROC eğrisi ve AUC hesaplıyoruz
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_binarized[:, i], y_pred_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# ROC eğrisini çiziyoruz
colors = cycle(['blue', 'red', 'green'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label='ROC curve of class {0} (area = {1:0.2f})'.format(i, roc_auc[i]))

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multi-class ROC')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.tree import plot_tree
plt.figure(dpi=100)
plot_tree(pipe_model["DT_model"], filled=True, fontsize=8, max_depth=7);
```
